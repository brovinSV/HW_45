```python
"""
Використовуючи модель ARIMA і датасет Yearly sunspots data 1700-2008 побудуйте прогноз на 2009-2030 р.р.
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.api import datasets
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
import statsmodels.api as sm
```

```python
# Завантаження датасету Yearly sunspots data 1700-2008
data = datasets.sunspots.load_pandas().data
data
```

```python
# Перетворення даних у тимчасовий ряд
ts = pd.Series(data['SUNACTIVITY'].values, index=data['YEAR'])
```

```python
# Побудова графіка часового ряду
plt.plot(ts)
plt.xlabel('Year')
plt.ylabel('Sunspot Activity')
plt.title('Yearly Sunspot Activity')
plt.show()
```

```python
# Перевірка стаціонарності часового ряду
def check_stationarity(timeseries):
    # Обчислення ковзної статистики та критичних значень
    rolmean = timeseries.rolling(window=12).mean()
    rolstd = timeseries.rolling(window=12).std()
    
    # Побудова графіка ковзної статистики та критичних значень
    plt.plot(timeseries, label='Original')
    plt.plot(rolmean, label='Rolling Mean')
    plt.plot(rolstd, label='Rolling Std')
    plt.xlabel('Year')
    plt.ylabel('Sunspot Activity')
    plt.title('Rolling Mean and Standard Deviation')
    plt.legend()
    plt.show()
    
     # Проведення тесту Дікі-Фуллера
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
# Перевірка стаціонарності часового ряду
check_stationarity(ts)
```

```python
# Виберіть стовпець "SUNACTIVITY" як ендогенну змінну
endog = data['SUNACTIVITY']
```

```python
# Будуємо графік автокореляційної функції
plot_acf(endog)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function')
plt.grid(True)
plt.show()
```

```python
# Будуємо графік часткової автокореляційної функції
plot_pacf(endog)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function')
plt.grid(True)
plt.show()
```

```python
# Створюємо модель SARIMAX і підгоняємо її до даних
model = sm.tsa.SARIMAX(endog, order=(5, 1, 0))
model_fit = model.fit()
```

```python
# Прогнозування на майбутні роки
future_years = pd.date_range(start='2009', end='2030', freq='Y')
forecast = model_fit.predict(start=len(ts), end=len(ts) + len(future_years)-1, typ='levels')
```

```python
# Графік прогнозу
plt.plot(ts, label='Observed')
plt.plot(future_years, forecast, label='Forecast')
plt.xlabel('Year')
plt.ylabel('Sunspot Activity')
plt.title('Yearly Sunspot Activity Forecast')
plt.legend()
plt.show()
```
